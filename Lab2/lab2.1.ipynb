```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
# Load dataset
df = pd.read_csv("House Price India.csv")
```

```python
# Use ONLY distance
X = df['Distance from the airport'].values
y = df['Price'].values
```

```python
# Train-test split (80-20)
n = len(X)
split = int(0.8 * n)

X_train = X[:split]
X_test = X[split:]
y_train = y[:split]
y_test = y[split:]

# Training Manual
x_mean = np.mean(X_train)
y_mean = np.mean(y_train)

numerator = np.sum((X_train - x_mean) * (y_train - y_mean))
denominator = np.sum((X_train - x_mean) ** 2)

b1 = numerator / denominator
b0 = y_mean - b1 * x_mean
```

```python
# Testing
y_pred = b0 + b1 * X_test
```

```python
# Least Square Error
lse = np.sum((y_test - y_pred) ** 2)

# RMSE (Root Mean Squared Error)
rmse = np.sqrt(np.mean((y_test - y_pred) ** 2))

# R² Score (Manual)
ss_total = np.sum((y_test - np.mean(y_test)) ** 2)
ss_residual = np.sum((y_test - y_pred) ** 2)
r2 = 1 - (ss_residual / ss_total)
```

```python
print("Simple Linear Regression (Manual)")
print("b0 (Intercept):", b0)
print("b1 (Slope):", b1)
print("Least Square Error (Test):", lse)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python
plt.figure()

# Scatter of actual data
plt.scatter(X_test, y_test, color='blue')

# Regression line
x_line = np.linspace(X_test.min(), X_test.max(), 100)
y_line = b0 + b1 * x_line

plt.plot(x_line, y_line, color='red')

plt.xlabel("Distance from airport")
plt.ylabel("Price")
plt.title("Simple Linear Regression")
plt.show()
```
